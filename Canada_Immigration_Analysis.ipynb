## Question 3  ( 10 pts)
Data aggregation. Find the average number of immigrants for each area in a given year. 

Write a function named **mean_area** that, given a year between 1980 and 2013, return a series of the average number of immigrants for each area. 

```python
def mean_area(year):
    # Check if the given year is in the DataFrame
    if year not in df.columns:
        raise ValueError(f"Year {year} is not in the DataFrame")
    
    # Group the data by area and calculate the average number of immigrants for the given year
    avg_immigrants = df.groupby('AreaName')[year].mean()
    
    return avg_immigrants
```

## Question 4  ( 15 pts)
Find the top 3 origins withm maximum number of immigrants for each area. 

Write a function named **get_top3** that, given a year between 1980 and 2013, returns the records of top 3 origins with the maximum number of immigrants for each area. 

```python
def get_top3(year):
    if year not in df.columns:
        return None

    df_year = df[['OdName', 'AREA', 'AreaName', 'REG', 'RegName', 'DEV', 'DevName', year]].copy()

    # Group by AREA and sum up the immigrants
    df_year = df_year.groupby(['OdName','AREA', 'AreaName','REG', 'RegName', 'DevName'], as_index=False).sum()

    # Sort by AREA and the number of immigrants in descending order
    df_year = df_year.sort_values(['AREA', year], ascending=[True, False])

    # Get the top 3 immigrants for each AREA
    df_top3 = df_year.groupby('AREA').head(3)

    return df_top3.reset_index(drop=True)
```

```python
# sample code to test your answer
get_top3(2010)
```

## Question 5  ( 15 pts)
Create a time series named **reg_time_series** for each region. The index should be the years from 1980 to 2013. The columns should be the region names. And the values should be the total number of immigrants of each region.  

```python
# Create an empty DataFrame with the years as the index and regions as columns
reg_time_series = pd.DataFrame(index=range(1980, 2013), columns=df['RegName'].unique())

# Loop over the years and regions to fill in the DataFrame with total number of immigrants
for year in range(1980, 2014):
    for region in reg_time_series.columns:
        reg_time_series.loc[year, region] = df[(df['RegName'] == region) & (df[year] > 0)][year].sum()

# Name the index
reg_time_series.index.name = 'year'
```

## Question 6  ( 5 pts)
Plot the time series of **reg_time_series**. Write a function called **plot_time_series**, given regions, start year and end year, draw the line plot to display the number of immigrants of the selected regions during the time interval.

```python

import matplotlib.pyplot as plt

def plot_time_series(regions, start_year, end_year):
    df = reg_time_series.loc[start_year:end_year, regions]
    df.plot()
    plt.xlabel('Year')
    plt.ylabel('Number of Immigrants')
    plt.title('Immigration from Selected Regions')
    plt.show()
```

```python
# sample code to test your answer
plot_time_series(['Eastern Asia','Eastern Europe'], 1980, 2001)
```
